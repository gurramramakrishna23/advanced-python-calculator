# Python Calculator

This is a simple menu-driven calculator built using Python.
The project is created and executed in Google Colab.


## Features
- Addition
- Subtraction
- Multiplication
- Division
- Error handling for division by zero
- Continuous execution using loops
## Upgraded Features
- Power and Modulus operations
- Square Root calculation
- Use of Python math library
- Improved input validation
- Clean and user-friendly menu system


## How to Run
1. Run all the cells
2. Select an operation from the menu
3. Enter two numbers
4. View the result


```python
import math

# Basic operations
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: Division by zero is not allowed."
    return a / b

# Advanced operations
def power(a, b):
    return a ** b

def modulus(a, b):
    return a % b

def square_root(a):
    if a < 0:
        return "Error: Cannot calculate square root of a negative number."
    return math.sqrt(a)


def calculator():
    while True:
        print("\n===== Advanced Python Calculator =====")
        print("1. Addition")
        print("2. Subtraction")
        print("3. Multiplication")
        print("4. Division")
        print("5. Power")
        print("6. Modulus")
        print("7. Square Root")
        print("8. Exit")

        choice = input("Enter your choice (1-8): ")

        if choice == '8':
            print("Calculator exited successfully.")
            break

        try:
            if choice == '7':
                num = float(input("Enter number: "))
                result = square_root(num)

            elif choice in ['1', '2', '3', '4', '5', '6']:
                num1 = float(input("Enter first number: "))
                num2 = float(input("Enter second number: "))

                if choice == '1':
                    result = add(num1, num2)
                elif choice == '2':
                    result = subtract(num1, num2)
                elif choice == '3':
                    result = multiply(num1, num2)
                elif choice == '4':
                    result = divide(num1, num2)
                elif choice == '5':
                    result = power(num1, num2)
                elif choice == '6':
                    result = modulus(num1, num2)
            else:
                print("Invalid choice. Please select a valid option.")
                continue

        except ValueError:
            print("Invalid input. Please enter numeric values.")
            continue

        print("Result:", result)


# Run the calculator
calculator()
```
